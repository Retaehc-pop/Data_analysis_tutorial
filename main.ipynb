# **Mechine Learning**

**LogisticRegression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_csv[features_column],data_csv[targets], test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import metrics

print(pd.crosstab(predictions,y_test.values.ravel(),margins=True,rownames=['Predict'],colnames=['         Real']))
print(metrics.classification_report(y_test.values.ravel(),predictions))
```

**LinearRegression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data_csv = pd.read_csv("data.csv", sep=";")
data_csv['school'] = data_csv['school'].map({'GP':1, 'MS':0})
data_csv['sex'] = data_csv['sex'].map({'M':0,'F':1})
data_csv['address'] = data_csv['address'].map({'U':1,'R':0})
data_csv['age'] = data_csv['age'].map({15:4, 16:4, 17:3, 18:2, 19:1, 20:5, 21:0, 22:1})
data_csv['Medu'] = data_csv['Medu'].map({0:3, 1:0, 2:1, 3:0, 4:1})
data_csv['Fedu'] = data_csv['Fedu'].map({0:2, 1:0, 2:1, 3:0, 4:1})
data_csv['Mjob'] = data_csv['Mjob'].map({'at_home':0, 'health':2, 'other':1, 'services':1, 'teacher':1})
data_csv['Fjob'] = data_csv['Fjob'].map({'at_home':0, 'health':0, 'other':0, 'services':0, 'teacher':1})
data_csv['reason'] = data_csv['reason'].map({'course':0, 'home':1, 'other':1, 'reputation':1})
data_csv['guardian'] = data_csv['guardian'].map({'mother':1, 'father':1, 'other':0})
data_csv['traveltime'] = data_csv['traveltime'].map({1:0, 2:1, 3:1, 4:1})
data_csv['studytime'] = data_csv['studytime'].map({1:0, 2:1, 3:2, 4:2})
data_csv['failures'] = data_csv['failures'].map({0:0, 1:1, 2:2, 3:2})
data_csv['schoolsup'] = data_csv['schoolsup'].map({'no':1, 'yes':0})
data_csv['higher'] = data_csv['higher'].map({'no':0, 'yes':1})
data_csv['internet'] = data_csv['internet'].map({'no':0, 'yes':1})
data_csv['freetime'] = data_csv['freetime'].map({1:0, 2:2, 3:0, 4:1, 5:1})
data_csv['goout'] = data_csv['goout'].map({1:1, 2:2, 3:1, 4:0, 5:0})
data_csv['Dalc'] = data_csv['Dalc'].map({1:1, 2:0, 3:0, 4:1, 5:1})
data_csv['Walc'] = data_csv['Walc'].map({1:1, 2:1, 3:0, 4:0, 5:0})
data_csv['health'] = data_csv['health'].map({1:2, 2:0, 3:0, 4:0, 5:1})
```

```python
features_column = ['sex', 'age', 'address', 'Medu', 'Fedu', 'Mjob', 'Fjob',
                     'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup',
                     'higher', 'internet', 'freetime', 'goout',
                     'health','absences', 'Walc', 'Dalc','G1', 'G2']
target = ['G3']
```

```python
LR = LinearRegression(fit_intercept=True)
X_train, X_test, y_train, y_test = train_test_split(data_csv[features_column],data_csv[target], test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LR.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
x_real = np.linspace(0,20,100)
y_real = x_real

plt.scatter(y_test,predictions, c='r')
plt.plot(x_real, y_real, c='black', linewidth=2)
plt.title('Prediction')
plt.xlabel('Real')
plt.ylabel('Predicted')
print("The R^2 for the Dataset in train is: ", model.score(X_train,y_train))
print("The R^2 for the Datasets: ", model.score(X_test,y_test))
```

```python
MSE = np.sqrt(mean_squared_error(y_test, predictions))
MSE
```
